#Transforming Data Using a Function or Mapping


```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon","pastrami", "corned beef", "bacon","pastrami", "honey ham", "nova lox"],"bounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
```

```python
data["animal"] = data["food"].map(meat_to_animal)
data
```

```python
def get_animal(x):
     return meat_to_animal[x]

get_animal("bacon")
```

```python
data["food"].map(get_animal)
data
```

```python
data["food"].map(lambda x: meat_to_animal[x] if x in data["food"].values else np.nan)

```

#Discretization and Binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32,99]
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)
age_categories
```

```python
age_categories.codes
```

```python
age_categories.categories
```

```python
age_categories.categories[0]
```

```python
pd.value_counts(age_categories)
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]
pd.cut(ages,bins, labels=group_names)
```

```python
age_categories.categories[0]
```

```python
data = np.random.uniform(size=20)
pd.cut(data, 4, precision=2)
```

```python
data = np.random.binomial(n=10, p=0.5, size=20)  #imdi Python bu ii 20 kere yapyor, yani: 20 defa 10 kez yaz-tura atyor, Her defasnda ka kez yaz geldiini hesaplyor,Sonra bu 20 sonucu bir listeye koyuyor.
data
```
